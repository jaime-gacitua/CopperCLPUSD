```python
import pandas as pd
#import pandas.io.data as iod
from pandas_datareader import data
#Also import numpy and matplotlib. Because ....
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mp
mp.style.use('ggplot')
import ctypes 

import statsmodels.formula.api as sm
from pandas.tools.plotting import scatter_matrix

import openpyxl as xl
from datetime import datetime

import urllib.request as ul #url.request lib for handling the url
from bs4 import BeautifulSoup #bs for parsing the page
```

```python

def get_copper_data(url_copper):
    url_response=ul.urlopen(url_copper[0],timeout=5)
    web_copper = BeautifulSoup(url_response)
    tables_copper = web_copper.find_all('table')
    a = []
    data_copper = []
    holiday_flag = False
    for table in tables_copper:
        if(table.get('bgcolor') == '#ffffff'):
            for row in table.find_all('tr'):
                a = []
                holiday_flag = False
                for col in row.find_all('td'):
                    if col.get_text() == 'HOLIDAY' or col.get_text().find("Back") > 0 or \
                        col.get_text() == 'Holiday':
                        # print(col.get_text())
                        holiday_flag = True
                    aux_txt = col.get_text().replace(" ", "") 
                    aux_txt = aux_txt.replace("+", "")
                    aux_txt = aux_txt.replace("$", "")
                    aux_txt = aux_txt.replace("\n", "")
                    aux_txt = aux_txt.replace("August", "Aug")
                    aux_txt = aux_txt.replace("July", "Jul")
                    aux_txt = aux_txt.replace("June", "Jun")
                    aux_txt = aux_txt.replace("April", "Apr")
                    aux_txt = aux_txt.replace("Dec.", "Dec")
                    aux_txt = aux_txt.replace("Nov.", "Nov")
                    aux_txt = aux_txt.replace("Oct.", "Oct")
                    aux_txt = aux_txt.replace("Sept.", "Sep")
                    aux_txt = aux_txt.replace("Apr.", "Apr")
                    aux_txt = aux_txt.replace("Mar.", "Mar")
                    aux_txt = aux_txt.replace("Feb.", "Feb")
                    aux_txt = aux_txt.replace("Jan.", "Jan")
                    aux_txt = aux_txt.replace("Sept", "Sep")                    
                    a.append(aux_txt)
                a.append(url_copper[1])
                if len(a)>3 and not holiday_flag:
                    data_copper.append(a)
    return(data_copper)
```

```python
url_copper = []
for i in range(2, 10):
    url_copper.append(['http://www.tradeservice.com/copper_pricing/200'+str(i)+'.html', '200'+str(i)])
for i in range(0, 5):
    url_copper.append(['http://www.tradeservice.com/copper_pricing/201'+str(i)+'.html', '201'+str(i)])
url_copper.append(['http://www.tradeservice.com/copper_pricing/index.html', '2015'])

data_copper = []
for url_year in url_copper:
    data_copper.extend(get_copper_data(url_year))
print(len(data_copper))
```

```python
for i, row in enumerate(data_copper):
    for j in [2, 3, 4, 5]:
        try:
            data_copper[i][j] = float(data_copper[i][j])
        except:
            print("Error in " + str(i) + " " + str(j) +", value: " + str(data_copper[i][j]))
            data_copper[i][j] = None
            print("Error in " + str(i) + " " + str(j) +", value: " + str(data_copper[i][j]))
```

```python
data_dollar_excel = xl.load_workbook("DatosBCCH.xlsx", data_only=True)
data_dollar_excel_sheet = data_dollar_excel.get_sheet_by_name("DOLAR_OBS_ADO")
data_dollar = []
data_dollar_row = []
row = 15
while row < data_dollar_excel_sheet.get_highest_row():
    data_dollar_row =[]
    data_dollar_row.append(data_dollar_excel_sheet['J'+str(row)].value)
    data_dollar_row.append(data_dollar_excel_sheet['K'+str(row)].value)
    if data_dollar_row[0] != None:
        data_dollar.append(data_dollar_row)
    row = row + 1
```

```python
dollar_date = []
dollar_price = []
copper_date = []
copper_price = []

for row in data_dollar:
    dollar_date.append(row[0])
    dollar_price.append(row[1])

for row in data_copper:
    copper_date.append(row[1])
    copper_price.append(row[2])

#dupes = set([x for x in copper_date if copper_date.count(x) > 1])
#print(dupes)

#for d in dupes:
#    for row in data_copper:
#        if row[1] == d:
#            print(row)


dollar_ts = pd.Series(dollar_price, index=dollar_date, name="dollar_price")
copper_ts = pd.Series(copper_price, index=copper_date, name="copper_price")
data_all = pd.concat([dollar_ts, copper_ts], axis=1)

data_all.copper_price.plot(legend=True)
data_all.dollar_price.plot(secondary_y=True, legend=True, title="Copper price vs CLP/USD")

data_all.plot(kind='scatter', x='copper_price', y='dollar_price')


result = sm.ols(formula="dollar_price ~ copper_price", data=data_all).fit()
print(result.params)
print(result.summary())



#df_data_all_pct.plot(kind='scatter', x='copper_price', y='copper_price_sh'+str(j), 
#                                 title="Pct Change periods= "+str(i+1)+" Shift= "+str(j))
```
